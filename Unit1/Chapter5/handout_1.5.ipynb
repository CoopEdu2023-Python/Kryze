# while Loop | while 循环

1. Write an endless while loop

    写一个 while 的死循环

```python
while 1:
    print(1)
```

2. Write a while loop that loops 100 times

    写一个循环 100 次的 while 循环

```python
i = 0
while i < 100:
    print(1)
```

3. How many times does the code below loop?

    下方代码循环了多少次？

```python
i = 1
s = 0
while i < 100:
    i *= 2
    s += 1
print(s)
```

3. How many times does the code below loop?

    下方代码循环了多少次？

```python
i = 1
s = 0
while i < 100:
    while i < 100:
        i += 1
        s += 1
print(s)
```

4. How many times does the code below loop?

    下方代码循环了多少次？

```python
i = 1
j = 1
s = 0
while i < 100:
    while j < 100:
        i += 1
        s += 1
    j += 1
print(s)
```

# for Loop | for 循环

5. Write a endless for loop

    写一个 for 的死循环

```python
for i in range(1, 5, -1):
    print(1)
```

6. Write a for loop that loops 100 times

    写一个循环 100 次的 for 循环

```python
s = 0
for i in range(100):
    s += 1
print(s)
```

7. How many times does the code below loop?

    下方代码循环了多少次？

```python
s = 0
for i in range(100):
    print("*", end='')
    s += 1
print(s)
```

8. How many times does the code below loop?

    下方代码循环了多少次？

```python
s = 0
for i in range(100):
    for j in range(100):
        print("*", end='')
        s += 1
print(s)
```
